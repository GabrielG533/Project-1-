```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

#Reading in teh cleaned data file
data = pd.read_csv('Non-Majors-Survey-Results(Cleaned).csv')

print("Column names in the dataset:", data.columns)

#Counting the gender values
gender_counts = data['gender'].value_counts()  

#Plotting the bar charts
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color='skyblue')
plt.title("Gender Distribution in CCM Non-Majors Survey")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)  

plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

#Reading in the clean data file
data = pd.read_csv('Non-Majors-Survey-Results(Cleaned).csv')

print("Column names in the dataset:", data.columns)

# Counting the occurrences of each race/ethnicity
# Replace 'race_ethnicity' with the actual column name if different
race_counts = data['race/ethnicity'].value_counts()  

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(race_counts.index, race_counts.values, color='lightgreen')
plt.title("Race/Ethnicity Distribution in CCM Non-Majors Survey")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Number of Students")
plt.xticks(rotation=45) 

plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

#Reading in the clean data file 
data = pd.read_csv('Non-Majors-Survey-Results(Cleaned).csv')

print("Column names in the dataset:", data.columns)

# Counting the occurrences of each age group
# Replace 'age' with the actual column name if different
age_counts = data['age_'].value_counts()  

age_counts = age_counts.sort_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(age_counts.index.astype(str), age_counts.values, color='orange')
plt.title("Age Distribution in CCM Non-Majors Survey")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)  


plt.show()
```
